```python
import pandas as pd
```

```python
data = pd.read_csv("tmdb_5000_movies.csv")
```

```python
m = 1000 #minimum votes
C = data["vote_average"].mean() #rating mean

data["WR"] = (data["vote_count"]/(data["vote_count"] + m)) * data["vote_average"] + (m/(data["vote_count"] + m)) * C
```

```python
data.sort_values(["WR"], ascending = False)[0:10]
```
